```python
import pandas as pd
import ast
import math
```

```python
filepaths = [
    'data/prob/prob_20/fake_news_20.csv',
    'data/prob/prob_40/fake_news_40.csv',
    'data/prob/prob_60/fake_news_60.csv',
    'data/prob/prob_80/fake_news_80.csv',
    'data/prob/prob_100/fake_news_100.csv',
    'data/prob/prob_20/xl_sum_20.csv',
    'data/prob/prob_40/xl_sum_40.csv',
    'data/prob/prob_60/xl_sum_60.csv',
    'data/prob/prob_80/xl_sum_80.csv',
    'data/prob/prob_100/xl_sum_100.csv'
]
```

```python
def clean_dataframe(df):
    def ensure_int_list(val):
        # Handle NaN
        if val is None or (isinstance(val, float) and math.isnan(val)):
            return []

        # Convert stringified list, e.g. "[1.0, 2.0]"
        if isinstance(val, str):
            try:
                val = ast.literal_eval(val)
            except Exception:
                return []  # invalid string → treat as empty

        # If it's a single float or int → wrap into list
        if isinstance(val, (int, float)):
            return [int(val)]

        # If it's a list → cast all elements to int
        if isinstance(val, list):
            return [int(float(x)) for x in val if x is not None and not (isinstance(x, float) and math.isnan(x))]

        # Otherwise return empty
        return []

    df["rand_word_idx"] = df["rand_word_idx"].apply(ensure_int_list)
    df["rand_sent_idx"] = df["rand_sent_idx"].apply(ensure_int_list)
    return df
```

```python
for path in filepaths:
    df = pd.read_csv(path)
    df = clean_dataframe(df)
    df.to_csv(path, index=False)
```
